```python
import pandas as pd
import csv
def TabelaExcel():
    ficheiro=pd.read_csv('diabetes_prediction_dataset.csv')
    print(ficheiro)

tabela=[]

def lerinformação():
    ficheiro = open('diabetes_prediction_dataset.csv', mode='r')
    inftabela= csv.reader(ficheiro,delimiter=',')
    nlinha=0
    for linha in inftabela:
        if nlinha==0:
            nlinha+=1
        else:
            tabela.append((linha[0],float(linha[1]),int(linha[2]),int(linha[3]),linha[4],float(linha[5]),float(linha[6]),int(linha[7]),int(linha[8])))

def distribuiçãoSexo():
    lerinformação()
    contadorM=0
    contadorF=0
    for t in tabela:
        sexo,_,_,doenca_cardiaca,_,_,_,_,_=t
        if doenca_cardiaca==1 and sexo== "Male":
            contadorM+=1
        elif doenca_cardiaca==1 and sexo=="Female":
            contadorF+=1
    print("Contagem masculina:", contadorM)
    print("Contagem feminina:", contadorF)

def distribuiçãoEtaria():
    lerinformação()
    escalões=[[20,24],[25,29],[30,34],[35,39],[40,44],[45,49],[50,54],[55,59],[60,64],[65,69],[70,74],[75,79],[80,84],[85,89]]
    lista=[]
    for t in tabela:
        _,idade,_,doenca_cardiaca,_,_,_,_,_=t
        for escalao in escalões:
            if doenca_cardiaca==1 and idade>=escalao[0] and idade <=escalao[1]:
                lista.append(escalao)
    for escalao in escalões:
        print("Escalão ",escalao," tem:",lista.count(escalao)," pessoas")


def distribuiçãoGlucose():
    lerinformação()
    lista=[]
    niveisGlucose=[]
    for i in range(0, 340,10):
            niveisGlucose.append([i,i+10])
    for t in tabela:
        _,_,_,doenca_cardiaca,_,_,_,nivel_glucose,_=t
        for nivel in niveisGlucose:
            if doenca_cardiaca==1 and nivel_glucose>=nivel[0] and nivel_glucose <=nivel[1]:
                lista.append(nivel)
    for nivel in niveisGlucose:
        print("Nivel", nivel, "tem:", lista.count(nivel), "niveis")

        

def tabelar():
    num=int(input("Escolha a distribuição que pretende tabelar:\n 1)sexo \n 2)idade \n 3)glucose"))
    if num==1:
        lerinformação()
        contadorM=0
        contadorF=0
        for t in tabela:
            sexo,_,_,doenca_cardiaca,_,_,_,_,_=t
            if doenca_cardiaca==1 and sexo== "Male":
                contadorM+=1
            elif doenca_cardiaca==1 and sexo=="Female":
                contadorF+=1
        print("Sexo masculino       |    Sexo Feminino        ")
        print("   ",contadorM,"                  ", contadorF)
    elif num==2:
        lerinformação()
        escalões=[[20,24],[25,29],[30,34],[35,39],[40,44],[45,49],[50,54],[55,59],[60,64],[65,69],[70,74],[75,79],[80,84],[85,89]]
        lista=[]
        for t in tabela:
            _,idade,_,doenca_cardiaca,_,_,_,_,_=t
            for escalao in escalões:
                if doenca_cardiaca==1 and idade>=escalao[0] and idade <=escalao[1]:
                    lista.append(escalao)
        print(" Escalão      |   Idade    ")
        for escalao in escalões:
            print("   ",escalao,"   ", idade)
    else:
        lerinformação()
        lista=[]
        niveisGlucose=[]
        for i in range(0, 340,10):
                niveisGlucose.append([i,i+10])
        for t in tabela:
            _,_,_,doenca_cardiaca,_,_,_,nivel_glucose,_=t
            for nivel in niveisGlucose:
                if doenca_cardiaca==1 and nivel_glucose>=nivel[0] and nivel_glucose <=nivel[1]:
                    lista.append(nivel)
        print(" Nivel  |  Pessoas")
        for nivel in niveisGlucose:
            print ( nivel, "  ",lista.count(nivel))

def menu():
    opção= int(input("introduza 1 nº entre 0 e 5: 0-TabelaExcel(); 1-lerinformação(); 2-distribuiçãoSexo(); 3-distribuiçãoEtaria(); 4-DistribuiçãoGlucose(); 5-Tabelar()"))
    if opção==0:
        TabelaExcel()
    elif opção==1:
        lerinformação()
    elif opção==2:
        distribuiçãoSexo()
    elif opção ==3:
        distribuiçãoEtaria()
    elif opção==4:
        distribuiçãoGlucose()
    elif opção==5:
        tabelar()

menu()


```
