```python
#app#
import random
menu=("Menu: \n(1) Criar Lista \n(2) Ler Lista \n(3) Soma \n(4) Média \n(5) Maior \n(6) Menor \n(7) estaOrdenada por ordem crescente \n(8) estaOrdenada por ordem decrescente \n(9) Procura um elemento \n(0) Sair")
print(menu)
lista=[]
sistem=True
while sistem==True:
    n=int(input("Menu: \n(1) Criar Lista \n(2) Ler Lista \n(3) Soma \n(4) Média \n(5) Maior \n(6) Menor \n(7) estaOrdenada por ordem crescente \n(8) estaOrdenada por ordem decrescente \n(9) Procura um elemento \n(0) Sair"))
    if n==1:
        b=int(input("Que tamanho vai ter a lista?"))
        while b>0:
            lista.append(random.randint(1,100))
            b=b-1
        print(lista)
        print(menu)
        
    if n==2:
        c=int(input("Que tamanho vai ter a lista?"))
        for i in range(1,c+1):
            num=int(input("intoduza um numero:"))
            lista.append(num)
        print(lista) 
        print(menu) 
    if n==3:
        if lista==[]:
            print("Não há lista")
            print(menu)
        else:
            soma=0
            for i in lista:
                soma=soma+i
                print("A soma dos números em ",lista,"é igual a",soma)
                print(menu)
    if n==4:
        if lista==[]:
            print("Não há lista")
            print(menu)
        else:
            soma1=0
            divisor=len(lista)
            for i in lista:
                soma1=soma1+1
            media=soma1/divisor
            print("A média dos números em ",lista," é ",media)
            print(menu) 
    if n==5:
        if lista==[]:
            print("Não há lista")
            print(menu)
        else:
            maior=lista[0]
            for i in lista:
                if i>maior:
                    i=maior
            print("O maior numero em ",lista," é ",maior)
            print(menu)
    if n==6:
        if lista==[]:
            print("Não há lista")
            print(menu)
        else:
            menor=lista[0]
            for i in lista:
                if i<menor:
                    i=menor
            print("O menor numero em ",lista," é ",menor)
            print(menu)
    if n==7:
        if lista==[]:
            print("Não há lista")
            print(menu)
        else:
            listaCres=[]
            for i in lista:
                listaCres.append(i)
            listaCres.sort()
            if listaCres==lista:
                print("Sim")
            if listaCres!= lista:
                print("Não")
            print(menu)
    if n==8:
        if lista==[]:
            print("Não há lista")
            print(menu)
        else:
            listaDecres=[]
            for i in lista:
                listaDecres.append(i)
            listaDecres.sort(reverse=True)
            if listaDecres==lista:
                print("Sim")
            if listaDecres!=lista:
                print("Não")
            print(menu)    
    if n==9:
        if lista==[]:
            print("Não há lista")
            print(menu)
        numeroPretendido=int(input("Qual o numero que pretende procurar?"))
        if numeroPretendido in lista:
            index=lista.index(numeroPretendido)
            print("O número pretendido,{} em {} está na posição {}".format(numeroPretendido, lista,index))
            print(menu)
        if numeroPretendido not in lista:
            print("{} não existe na lista {}".format(numeroPretendido,lista))
            print(menu)
    if n==0:
        print("## See yah ##")
        sistem=False
 
```
